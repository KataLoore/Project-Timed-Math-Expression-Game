```python
#@title Timed Math-Expression Quiz
'''
This maths game challenges the player to answer a set amount of randomly
generated three-operand expressions connected with randomly picked operators
from a set of three options *op*.
'''
import random
import time

min = 3
max = 10
op = ['+', '-', '*']
count = 3

def random_problem():
  v1 = random.randint(min, max)
  v2 = random.randint(min, max)
  v3 = random.randint(min, max)
  o1 = random.choice(op)
  o2 = random.choice(op)

  expression = str(v1) + ' ' + o1 + ' ' + str(v2) + ' ' + o2 + ' ' + str(v3)
  answer = eval(expression)

  return expression, answer

def Play():
  wrong_count = 0
  print('Game starting in:')
  time.sleep(1)
  print("3")
  time.sleep(1)
  print("2")
  time.sleep(1)
  print("1")
  time.sleep(1)
  print("\033[1mGo!\033[0m")
  time.sleep(1)
  print('')

  start_time = time.time()

  i = 0
  while i < count:
    expression, answer = random_problem()
    while True:

      print('Question ' + str(i + 1) + ': ')
      guess = input(expression + ' = ')
      if guess == str(answer):
        print('\033[32mCorrect!\033[0m')
        print('-----------')
        i += 1
        break
      elif guess == 'n':
        print('\033[33mSkipping expression...\033[0m')
        print('-----------')
        wrong_count += 1
        break
      print('\033[31mWrong, try again!\033[0m')
      print('-----------')
      wrong_count += 1

  end_time = time.time()
  total_time = round(end_time - start_time, 2)

  print('')
  time.sleep(1)
  print('-----------------------------')
  print('| Your stats:               |')
  print('-----------------------------')
  time.sleep(1)
  print(' - Total time: ' + str(total_time) + ' seconds')
  time.sleep(1)
  print(' - Wrong answers: ' + str(wrong_count))
  print('-----------------------------')
  print('')
  time.sleep(2)

print("\033[1mTimed Math Game\033[0m")
print('')
time.sleep(1)
print('- Correctly answer a total of ' + str(count) + ' expressions in the shortest amount of time.')
time.sleep(2)
print('- Stuck? To skip foreward to the next expression, type "n" as your answer')
time.sleep(2)
print('')
print('Press enter to start...')
input('')

playing = True
while playing:
  Play()
  print('\033[1mPlay again?\033[0m')
  time.sleep(1)
  print('Yes: type "\033[32my\033[0m" | No: type "\033[31mx\033[0m"')
  replay = input('')
  if replay == 'y':
    Play()
  elif replay == 'x':
    print('')
    print('\033[1mThanks for playing!\033[0m')
    playing = False
```
